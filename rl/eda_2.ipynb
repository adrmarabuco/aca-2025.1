### epsilon greedy

```python
files = [
("evaluation_epsilon_greedy_exp_30_05.csv", "mode = exponential, alpha = 0.1, gamma = 1.0, min_epsilon = 0.1, max_epsilon = 0.9, k = 0.05, max_time=30"),
("evaluation_epsilon_greedy_exp_30_1.csv", "mode = exponential, alpha = 0.1, gamma = 1.0, min_epsilon = 0.1, max_epsilon = 0.9, k = 0.1, max_time=30"),
("evaluation_epsilon_greedy_exp_30_2.csv", "mode = exponential, alpha = 0.1, gamma = 1.0, min_epsilon = 0.1, max_epsilon = 0.9, k = 0.2, max_time=30"),
("evaluation_epsilon_greedy_fx_2_30.csv", "mode = fixed, alpha = 0.1, gamma = 1.0, max_epsilon = 0.2, max_time=30"),
("evaluation_epsilon_greedy_fx_5_30.csv", "mode = fixed, alpha = 0.1, gamma = 1.0, max_epsilon = 0.5, max_time=30"),
("evaluation_epsilon_greedy_fx_8_30.csv", "mode = fixed, alpha = 0.1, gamma = 1.0, max_epsilon = 0.8, max_time=30"),
("evaluation_epsilon_greedy_lin_30.csv", "mode = linear, alpha = 0.1, gamma = 1.0, min_epsilon = 0.1, max_epsilon = 0.9, max_time=30")
]
```

```python
dfs = []
for file in files:
    tmp = pd.read_csv(file[0])
    tmp['name'] = file[0][:-4]
    tmp['args'] = file[1]
    dfs.append(tmp)

df = pd.concat(dfs)
df = df.reset_index(drop=True)
```

```python
df.sort_values(by="mean", ascending=False)
```

```python
df.sort_values(by="mean", ascending=False).round(2).to_latex('results2.tex')
```

```python
sns.heatmap(df.pivot(index='alpha', columns='gamma', values='mean'), annot=True, cmap='RdYlBu')
```

### UCB

```python
files = [
("evaluation_upper_confidence_bound_05_30.csv", "mode = fixed, alpha = 0.1, gamma = 1.0, c = 0.5, max_time=30"),
("evaluation_upper_confidence_bound_1_30.csv", "mode = fixed, alpha = 0.1, gamma = 1.0, c = 1, max_time=30"),
("evaluation_upper_confidence_bound_2_30.csv", "mode = fixed, alpha = 0.1, gamma = 1.0, c = 2, max_time=30"),
("evaluation_upper_confidence_bound_3_30.csv", "mode = fixed, alpha = 0.1, gamma = 1.0, c = 3, max_time=30"),
]
```

```python
dfs = []
for file in files:
    tmp = pd.read_csv(file[0])
    tmp['name'] = file[0][:-4]
    tmp['args'] = file[1]
    dfs.append(tmp)

df = pd.concat(dfs)
df = df.reset_index(drop=True)
```

```python
df.sort_values(by="mean", ascending=False).head()
```

```python
sns.heatmap(df.pivot(index='alpha', columns='gamma', values='std'), annot=True, cmap='RdYlBu_r')
```

```python
fig, ax = plt.subplots(2, 2, figsize=(12, 12))
sns.regplot(data=df, x="alpha", y="std", ax=ax[0, 0])
sns.regplot(data=df, x="alpha", y="mean", ax=ax[0, 1])
sns.regplot(data=df, x="gamma", y="std", ax=ax[1, 0])
sns.regplot(data=df, x="gamma", y="mean", ax=ax[1, 1])
```
